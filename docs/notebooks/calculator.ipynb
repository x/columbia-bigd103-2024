```python
street = int(input("What street are you on?"))

if street > 200:
    print("You're in the Bronx")
elif street >= 116 and street <= 120:
    print("You're at Columbia")
elif street > 56:
    print("You're uptown")
elif street > 14:
    print("You're midtown")
else:
    print("You're downtime")
```
